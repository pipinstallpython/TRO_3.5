#### Условие задачи:
У вас есть шарик массой 0,1 кг, который находится на высоте 2 метра над землей. Вы отпускаете шарик, и он начинает свободно падать под действием силы тяжести. Используя законы движения свободного падения, постройте график зависимости скорости шарика от времени за первые 4 секунды его падения.
Помните, что скорость свободно падающего тела увеличивается со временем с постоянным ускорением, равным ускорению свободного падения, приблизительно равному 9,8 м/с².

#### Решение:
Для решения этой задачи мы можем использовать закон движения свободного падения, который гласит:
$$ h = \frac{gt^{2}}{2} $$ 
где $h$ - высота, с которой падает тело, $g$ - ускорение свободного падения (приблизительно 9,8 м/с²), $t$ - время падения.

Найдем скорость падающего тела:
$$ v = gt $$
где v - скорость, g - ускорение свободного падения, t - время падения.

```python
import matplotlib.pyplot as plt
```

```python
h = 2 
m = 0.1
g = 9.8 
t_max = 4 
```

```python
time = []
velocity = []
```

```python
for t in range(0, t_max+1):
    time.append(t)
    v = g*t
    velocity.append(v)
```

Построим график зависимости скорости шарика от времени за первые 4 секунды его падения:
На оси X графика будет откладываться время в секундах, а на оси Y - скорость в метрах в секунду.

```python
# Строим график зависимости скорости от времени
plt.plot(time, velocity)
plt.title('Зависимость скорости шарика от времени')
plt.xlabel('Время, с')
plt.ylabel('Скорость, м/с')
plt.grid()
plt.show()
```

На графике видно, что скорость шарика увеличивается со временем с постоянным ускорением, равным ускорению свободного падения. За первые 4 секунды падения скорость шарика увеличивается с 0 до приблизительно 39,2 м/с.
