#### Условие задачи:
Цель данной задачи заключается в разработке модели прогнозирования цен на недвижимость на основе ее характеристик. У нас имеется набор данных, включающий информацию о площади и количестве комнат для каждого объекта недвижимости, а также соответствующие им цены.

Для достижения поставленной цели мы планируем использовать метод линейной регрессии, который позволяет моделировать линейную зависимость между независимыми переменными (площадь и количество комнат) и зависимой переменной (цена недвижимости).

#### Решение:
Шаги для решения задачи прогнозирования цен на недвижимость на основе характеристик:

1) Загрузка данных о недвижимости.
2) Подготовка данных (обработка выбросов, заполнение пропущенных значений).
3) Разделение данных на обучающий и тестовый наборы.
4) Обучение модели линейной регрессии на обучающем наборе данных.
5) Оценка производительности модели на тестовом наборе данных (с использованием метрик, например, MAE или R^2).
6) Прогнозирование цен на новую недвижимость с использованием обученной модели.
7) Визуализация результатов (например, построение графика фактических и предсказанных цен).

Каждый из этих шагов требует использования соответствующих библиотек и написания кода на Python, таких как pandas, scikit-learn и matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
data = pd.read_excel('sam_7.xlsx')

# Извлечение значений из загруженный данных
area = data['Area'].values
rooms = data['Rooms'].values
price = data['Price'].values
```

```python
X = np.column_stack((area, rooms))
```

```python
model = LinearRegression()
model.fit(X, price);
```

Предсказание цен на новые недвижимости

```python
new_area = np.array([120, 250, 350, 400, 500])
new_rooms = np.array([3, 4, 5, 6, 7])
new_data = np.column_stack((new_area, new_rooms))
predicted_prices = model.predict(new_data)
```

График прогнозирования цен на недвижимость

```python
plt.scatter(area, price, color='blue', label='Фактические данные')
plt.plot(new_area, predicted_prices, color='red', label='Прогнозируемые данные')
plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.legend()
plt.show()
```

Из этого графика мы можем видеть взаимосвязь между площадью недвижимости и ее ценой. Синие точки представляют фактические данные о недвижимости, где по оси X отложена площадь, а по оси Y - цена. Красная линия представляет предсказанные цены для новых недвижимостей с заданными характеристиками площади и количества комнат.
