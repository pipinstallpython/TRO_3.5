### Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
import requests
```

```python
from io import BytesIO
```

```python
plt.figure(figsize=(20, 20))

url = requests.get('https://phonoteka.org/uploads/posts/2021-07/1625504295_10-phonoteka-org-p-skandinavskaya-mifologiya-na-rabochii-stol-10.jpg')
img = Image.open(BytesIO(url.content))

plt.imshow(img);
```
