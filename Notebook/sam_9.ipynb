#### Условие:
Компания производит и продает три вида продукции: книги, музыкальные инструменты и спортивное оборудование. Проанализируйте, какой процент общего дохода компании приходится на каждый вид продукции, чтобы определить, на какую продукцию следует сосредоточить больше внимания в бизнес-стратегии.

Общий доход компании за прошлый год составил 1 000 000 долларов США. Ниже приведена таблица, показывающая, какой доход был получен от каждого вида продукции:
![1](https://sun9-54.userapi.com/impg/qAJoHZPm15iX8HC-ifUWNXO0aEaRhSsAyziPBg/4f8E34GbOq4.jpg?size=690x174&quality=96&sign=2a1978d72689ef5c9f4d94b7e0b9ae59&type=album)

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Книги', 'Музыкальные инструменты', 'Спортивное оборудование']
sizes = [300000, 450000, 250000]
```

```python
colors = ['lightpink', 'MediumPurple', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Доход компании по видам продукции')
plt.axis('equal')
plt.show()
```

Данная круговая диаграмма показывает какой процент общего дохода компании приходится на каждый вид продукции. Как видно из диаграммы, большую часть дохода компании (45%) приносят музыкальные инструменты, в то время как книги приносят лишь 30% от общего дохода.
